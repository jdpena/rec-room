```python
from rec_room.datasets import load_data
from rec_room import ROOT_DIR
from os.path import join
```

```python
def create_us_colleges_dataset():
    d4 = load_data('school_and_country_table.csv')
    #d4.pop('country')
    #d4 = d4[:401] 
    d4 = d4.loc[d4['country'] == 'United States of America'].reset_index(drop=True)
    _ = d4.pop('country')
    fpath = join(ROOT_DIR, 'datasets', 'data')
    fname = "us_college_names.csv"
    f = join(fpath, fname)
    ################# d4.to_csv(f, index=False)
    return pd.read_csv(f)
d4 = create_us_colleges_dataset()
d4
```

```python
# rank: 1 high - 4 low
d1 = load_data('college_rank.csv')
_ = d1.pop('admit')
d1
```

```python
d2 = d1.sample(n=len(d4))
d2.reset_index(drop=True)
```

```python
ones = d2[d2.loc[:, 'rank'] == 1].index
twos = d2[d2.loc[:, 'rank'] == 2].index
threes = d2[d2.loc[:, 'rank'] == 3].index
fours = d2[d2.loc[:, 'rank'] == 4].index
n1, n2, n3, n4 = len(ones), len(twos), len(threes), len(fours)
print(n1,n2,n3,n4)
```

```python
uni_1 = d4[:n1]#.sample(n=n1) # , random_state = 1)
uni_2 = d4[n1:n1+n2]#.sample(n=n2) # , random_state = 1)
uni_3 = d4[n1+n2:n1+n2+n3]#.sample(n=n3) # , random_state = 1)
uni_4 = d4[n1+n2+n3:n1+n2+n3+n4]#.sample(n=n4) # , random_state = 1)
```

```python
d5 = d2.copy()
```

```python
d5.loc[ones, 'name'] = uni_1.values
d5.loc[twos, 'name'] = uni_2.values
d5.loc[threes, 'name'] = uni_3.values
d5.loc[fours, 'name'] = uni_4.values
```

```python
d5 = d5.reset_index(drop=True)
d5
```
